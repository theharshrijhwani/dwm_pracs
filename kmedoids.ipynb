# KMedoids

## Without SKLearn

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.randn(300,2)*2+np.array([[6,6]])

def assign_cluster(X, medoids):
    clusters = [[] for _ in range(len(medoids))]
    for point in X:
        distances = [np.linalg.norm(point-medoid) for medoid in medoids]
        cluster_idx = np.argmin(distances)
        clusters[cluster_idx].append(point)
    return [np.array(cluster) for cluster in clusters]

def kmedoids(X, k, max_iterations=100):
    medoids = X[np.random.choice(len(X), k, replace=False)]
    for _ in range(max_iterations):
        clusters = assign_cluster(X, medoids)
        new_medoids = [np.median(cluster,axis=0) for cluster in clusters]
        if all(np.array_equal(medoid, new_medoid) for medoid, new_medoid in zip(medoids, new_medoids)):
            break
        medoids = new_medoids
    return medoids, clusters

k = 3
medoids, clusters = kmedoids(X, k)

plt.figure(figsize=(8, 6))
for idx, cluster in enumerate(clusters):
    plt.scatter(cluster[:, 0], cluster[:, 1], label=f'Cluster {idx + 1}')

plt.scatter(np.array(medoids)[:, 0], np.array(medoids)[:, 1], marker='X', s=200, c='red', label='Medoids')
plt.title('K-medoids Clustering')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## With SKLearn

```python
import numpy as np
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
%matplotlib inline

X = np.random.rand(100,2)
k = 3

kmedoids = KMedoids(n_clusters=k)

kmedoids.fit(X)

medoids = X[kmedoids.medoid_indices_]
labels = kmedoids.labels_

plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', edgecolors='k')
plt.scatter(medoids[:,0], medoids[:,1], marker="X", color="red", s=200)
plt.title("K-Medoids Clustering")
plt.legend()
plt.show()
```
